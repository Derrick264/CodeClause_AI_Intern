```python
import pandas as pd
import numpy as np
from pandas_profiling import ProfileReport
import seaborn as sns
from scipy.stats import mode
from sklearn.preprocessing import LabelEncoder
import warnings
warnings.filterwarnings("ignore")
import os
import cv2
import glob
import os
from tkinter import Tk
from tkinter.filedialog import askopenfilename
```

```python
#This is a variable that will store the image data after it's loaded. 
#You can name it whatever you prefer; in this case, it's named "image."
#The cv2 is an alias for the OpenCV library, and imread stands for "image read."

image=cv2.imread(r"C:\Users\Derrick T\Downloads\SampleImg.jpg")
```

```python
#This code displays the image stored in the "image" variable with the window title "Derrick" and 
#waits for a key press indefinitely.

cv2.imshow("Derrick",image)
cv2.waitKey(0)
```

```python
#This code converts the image loaded with OpenCV from BGR color space to RGB color space and 
#then displays it using Matplotlib's plt.imshow().

import matplotlib.pyplot as plt
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
#converts the original color image stored in the "image" variable into a grayscale image and stores it in the "grey_img" variable.

grey_img=cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
# cv2.cvtColor : This is a function from the OpenCV library used for color space conversion.

grey_img
```

##### cv2.COLOR_BGR2GRAY : This is a constant representing the color space conversion from BGR (Blue, Green, Red) to grayscale. In a grayscale image, each pixel represents the brightness or intensity of the color, and it is a single channel image with shades of gray. The conversion reduces the image from multiple color channels to a single grayscale channel.

```python
plt.imshow(grey_img)
```

```python
inverted_img = 255 - grey_img
plt.imshow(inverted_img)
```

##### inverted_img: This variable stores the inverted version of the grayscale image. It inverts the pixel values by subtracting each pixel value from 255, effectively reversing the brightness values. Dark areas become light, and vice versa.

```python
blurred = cv2.GaussianBlur(inverted_img,(21,21),0)
plt.imshow(blurred)
```

##### blurred: This variable stores the result of the Gaussian blur applied to the inverted grayscale image.

##### cv2.GaussianBlur(...): This function from the OpenCV library performs Gaussian blurring on the input image. It takes the following parameters:
##### (21, 21): The size of the Gaussian kernel used for blurring. A larger kernel size results in stronger blurring.
##### 0: This parameter represents the standard deviation of the Gaussian kernel. A value of 0 means that OpenCV calculates the standard deviation based on the kernel size.

```python
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(grey_img,inverted_blurred,scale=256.0)
plt.imshow(pencil_sketch)
```

##### inverted_blurred: This variable stores the inverted version of the blurred image, similar to the previous inversion. It's done to revert the blurring effect while keeping the inverted colors.


##### pencil_sketch: This variable is used to create the pencil sketch effect by dividing the original grayscale image (grey_img) by the inverted and blurred image (inverted_blurred). The cv2.divide(...) function is used for this purpose. The scale parameter is set to 256.0 to control the intensity of the sketch effect.

```python
cv2.imshow("Derrick" , pencil_sketch)
cv2.waitKey(0)
```

##### These lines of code display the "pencil_sketch" image in a window with the title "Derrick" using the cv2.imshow(...) function and wait for a key press indefinitely using cv2.waitKey(0).
